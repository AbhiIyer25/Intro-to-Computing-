```python
marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}
for name, grades in marks.items():
    print(name, "got a", grades)
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Andy", "Amy", "James", "Jules", "Arthur"])
y = np.array([88, 66, 90, 55, 77])
font1 = {"family": "serif", "color": "black", "size": 20}
font2 = {"family": "serif", "color": "black", "size": 15}
plt.title("Grade Chart", fontdict = font1)
plt.xlabel("Student's Name", fontdict = font2)
plt.ylabel("Student's Grade", fontdict = font2)
plt.bar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([88, 66, 90, 55, 77])
students = np.array(["Andy", "Amy", "James", "Jules", "Arthur"])
font1 = {"family": "serif", "color": "black", "size": 20}
plt.title("Grade Chart", fontdict = font1)
plt.pie(y, labels = students)
plt.legend(title = "Student Name", bbox_to_anchor = (1.05, 1.0), loc = "best")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

font1 = {"family": "serif", "color": "navy", "size": 15}
color = np.array(["red", "orange", "yellow", "green", "blue", "indigo", "purple", "pink", "tan", "gold",
                       "grey", "teal", "turquoise"])

#This is the graph with multiple lines
x1 = np.array([1, 2, 3, 4])
y1 = np.array([1, 2, 3, 4])
y2 = np.array([1, 3, 5, 7])
y3 = np.array([1, 4, 7, 10])
plt.subplot(2, 3, 1)
plt.plot(x1, y1, x1, y2, x1, y3)
plt.title("Multiple Lines", fontdict = font1)

#This is the bar graph
xrows = np.array(["A", "B", "C", "D"])
ybars = np.array([25, 50, 10, 70])
plt.subplot(2, 3, 2)
plt.bar(xrows, ybars, width = 0.5, color = "teal")
plt.title("Bar Graph", fontdict = font1)

#This is the pie chart
xpie = np.array([10, 20, 30, 40])
ylabels = np.array(["John", "Dave", "Bill", "Carl"])
plt.subplot(2, 3, 3)
plt.pie(xpie, labels = ylabels, colors = color)
plt.title("Pie Chart", fontdict = font1)
plt.legend(title = "Names", bbox_to_anchor = (1.15, 1.0), loc = "upper left")

#This is the graph with grid
xgrid = np.array([1, 2, 3, 4, 5, 6])
ygrid = np.array([10, 7, 9, 10, 2, 5])
plt.subplot(2, 3, 4)
plt.plot(xgrid, ygrid, marker = 'o', color = 'navy')
plt.title("Graph with Grid", fontdict = font1)
plt.grid(color = "orange", linestyle = ':', linewidth = 2)

#This is the scatter plot
xpoints = np.array([20, 6, 16, 20, 13, 10, 5, 16, 9, 1, 3, 7, 8])
ypoints = np.array([90, 70, 88, 93, 85, 73, 70, 81, 75, 60, 66, 85, 84])
plt.subplot(2, 3, 5)
plt.scatter(xpoints, ypoints, c = color)
plt.title("Scatter Plot", fontdict = font1)

#This is the histogram plot
xamounts = np.array([1, 1.5, 2, 1, 4, 5, 2.5, 3.5, 6, 1.5, 1.5, 2.5, 3.5, 4, 3, 2, 2, 5, 4.5, 1.5, 4])
plt.subplot(2, 3, 6)
plt.hist(xamounts, color = "darkred")
plt.title("Histogram", fontdict = font1, )

plt.show()
```
