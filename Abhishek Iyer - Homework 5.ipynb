Question 2

```r
# Reading the Amazon Transactions
amazon = read.csv("amazon-orders.csv")
amazon
```

```r
# Removing the columns that we do not need and renaming the columns
# Stripping the tax and total amounts by getting rid of the "$" 
# Converting the Tax and Total columns to numbers, and Date columns to yyyy-dd-mm format
amazon_orders = amazon[c(1,20,21)]
colnames(amazon_orders) <- c('Date', 'Tax', 'Total')
amazon_orders$Total <- substring(amazon_orders$Total, 2)
amazon_orders$Tax <- substring(amazon_orders$Tax, 2)
amazon_orders$Tax <- as.numeric(as.character(amazon_orders$Tax))
amazon_orders$Total <- as.numeric(as.character(amazon_orders$Total))
amazon_orders$Date <- as.Date(as.character(amazon_orders$Date), format = "%m/%d/%y")
amazon_orders
```

```r
# Statistical Analysis for Amounts
mean(amazon_orders$Total)
median(amazon_orders$Total)
min(amazon_orders$Total)
max(amazon_orders$Total)
```

```r
# Analyzing which number purchase had the maximum and minimum spending
which.max(amazon_orders$Total)
which.min(amazon_orders$Total)
```

```r
# Counting off on the list and separating the data frame by month into 4 separate data frames
april <- amazon_orders[c(1:10), ]
may <- amazon_orders[c(11:28), ]
june <- amazon_orders[c(29:34), ]
july <- amazon_orders[c(35:59), ]

# Monthly summary of totals, tax amounts, and number of purchases per month
summary(april)
dim(april[1])
summary(may)
dim(may[1])
summary(june)
dim(june[1])
summary(july)
dim(july[1])
```

```r
# Bar Plot of Total Amount Spent per Month
x <- c('April', 'May', 'June', 'July')
y <- c(sum(april$Total), sum(may$Total), sum(june$Total), sum(july$Total))
colors <- c("hotpink", "navy", "gold", "aquamarine")
barplot(y, names.arg = x, col = colors, main = "Historical Transactions per Month")
```
